### **Introduction**
- #### Rust aik system programming language hai jo memory safety, fast performance, aur multi-threading (concurrency) ko secure tarike se handle karti hai.
- #### Iska goal hai: speed + safety bina overhead (zero-cost abstractions).
- #### Isko Graydon Hoare ne Mozilla Research mein banaya, aur Brendan Eich (JavaScript creator) ne bhi contribute kiya.
- #### Developers Rust ko is liye pasand kartay hain:
    - #### memory errors se bachata hai
    - #### bohot fast run hota hai
    - #### multi-threaded code safe banata hai
    - #### strong type system deta hai

### **Interpreter , Compiler Based ?**
- #### Rust pure compiler-based language hai.
- #### Compiler ka naam: rustc
- #### Aur build system + package manager ka naam: cargo
- #### Native machine code (binary) bana deta hai — bilkul C/C++ ki tarah.

### **rustc**
- #### **rustc** Rust ka compiler hy
    - #### e.g **rustc main.rs**
    - #### rustc compiler .rs file ko machine code (binary) mein convert karta hai.

### **main Function**

- #### Rust ka main function program ka starting point hota hai — bilkul C aur C++ ki tarah.
- #### Jab tum Rust ka program run karte ho, sabse pehle main() function hi execute hota hai.
- #### ; statement ko finish (terminate) karne ke liye lagaya jata hai.

````rust
fn main() {
    println!("Hello Rust!");
}
````

### **println! Macro**

- #### Macro ek special code hai jo compile time par expand hota hai.
- #### Matlab: compiler ke samay ye code generate karta hai, run time pe nahi
- #### println! Rust ka macro hai, jo console par output print karta hai..
- #### Note: ! matlab ye macro hai, function nahi.

````rust
fn main() {
    println!("Mubeen");
}
````

    <div style="color: black; font-size: 15px; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black; padding: 20px;">

### **{} placeholder**

- #### Rust mein println! ya format! jaisi macros mein use hota hai
- #### Ye value ko string mein insert karne ke liye hai

````rust
fn main() {
    println!("value of x is {}",2);
    println!("value of a is {} Value of b is {}",3,4);
}
````

### **Positional Arguments**

- #### Positional Arguments wo values hain jo placeholder {} ke order ke hisaab se assign hoti hain.
- #### Matlab {0} = pehla argument, {1} = doosra argument, {2} = teesra argument…
- #### Rust mein ye println! ya format! macros mein use hotay hain.

````rust
fn main() {
    println!("First Name : {1} Last Name {0}","Mubeen","Ahmad");
}
````

- #### Rust mein placeholder ke bina argument ignore karna allowed nahi.
- #### Agar extra argument hai → ya to use karo, ya remove karo





````rust
fn main() {
    println!("First Name : {0} Last Name {0}","Mubeen","Ahmad");
}
````

### **Named Arguments**

- #### Named Arguments = placeholders ke naam ke sath arguments assign karna
- #### Ye Positional Arguments se flexible aur readable hain

````rust
fn main() {
    println!("First Name : {fname} Last Name {lname}",fname="Mubeen",lname="Ahmad");
}
````
