### **What is Variable**

- #### Variable ek naam hota hai jo computer ki memory ki kisi jagah ko diya jata hai.
- #### Program ke andar data store karne ka yeh sabse basic tareeqa hota hai
    - #### Computer ki RAM me aik choti si jagah banti hai
    - #### Us jagah ko hum ek naam de dete hain → is naam ko variable kehte hain
    - #### Jab bhi program ko koi value store ya use karni hoti hai, wo us naam ke zariye memory tak pohanchta hai

```rust
fn main(){

    let name = "Mubeen";
    println!("{}",name);
    
}
```

### **Immutable Variable**

- #### Immutable ka matlab hota hai:
    - #### Ek baar value set hogayi, baad me change Nahi ho sakti.”
    - #### Yani variable lock ho jata hai — uski value fix rehti hai.
    - #### Rust me variables default immutable hote hain

<br>

- #### Benefits Immutable
    - #### Fast hota hai
    - #### Memory safe hota hai
    - #### Bugs kam hotay hain
    - #### Compiler zyada optimize karta hai (performance increase)

```rust
fn main(){
    let x = 10;
    x = 20;   // error — immutable variable ko change nahi kar sakte
}
```

    <div style="color: black; font-size: 15px; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black; padding: 20px;">

### **Mutable Variable**

- #### Mutable ka matlab hota hai:
    -  #### “Is variable ki value tum baad me change kar sakte ho.”
    -  #### Yani yeh changeable hota hai.
    -  #### Rust me variables default mutable nahi hote.
    -  #### Aapko specially mut lagana parta hai.
        -  #### mut ek keyword hai
        -  #### Ye variable mutable hai, iski value baad me change ho sakti hai.

```rust
fn main(){
    let mut x = 10;
    
    println!("x before {}",x);
    
    x = 20;
    
    println!("x after {}",x);
}
```

### **Shadowing**

- #### Shadowing ka matlab:
    - #### Ek naya variable create karna jo purane variable ke naam ke saath overlap karta hai.
    -  #### Purana variable immutable rehta hai
    -  #### Naya variable same naam se ban jata hai aur purane ko temporarily shadow kar deta hai
    -  #### Memory me naya jagah ban jaati hai, purana delete nahi hota

```rust
fn main(){
    
    let x = 5;         // x immutable
    println!("Value of x is {}",x);
    let x = x + 1;     // naya x, purana x ko shadow kar raha hai
    println!("Value of x Now is {}", x); // 6

}
```

- #### Pehla x = 5 immutable tha
- #### Dusra x = x + 1 → naya x bana gaya
- #### Purana x ab shadowed hai, compiler usko ignore karta hai
- #### Memory-wise: purana x delete nahi hota, lekin ab program me dusra x use hoga.

<body style="font-family: Sans-serif;">
    <div style="color: black; font-size: 20px; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black; padding: 20px;">

| Feature             | Immutable (`let`)              | Shadowing (`let` dobara)                                    | Mutable (`let mut`)                  |
| ------------------- | ------------------------------ | ----------------------------------------------------------- | ------------------------------------ |
| **Definition**      | Value change nahi ho sakti     | Naya variable create hota hai same naam se                  | Value change ho sakti hai            |
| **Value Change**    | ❌ Not allowed                  | ❌ Purana variable change nahi hota, naya value set hoti hai | ✔ Allowed                            |
| **Keyword**         | `let`                          | `let`                                                       | `let mut`                            |
| **Memory**          | Ek hi memory location          | Nayi memory location banti hai (shadow)                     | Same memory me value update hoti hai |
| **Use Case**        | Safe aur fast                  | Jab same naam se variable redefine karna ho                 | Loops, counters, state change        |
| **Example**         | `let x = 5;`                   | `let x = 5; let x = x + 1;`                                 | `let mut x = 5; x = 6;`              |
| **Error on change** | Hoga if try to modify `x = 6;` | ❌ No error, new variable created                            | ❌ No error, value update hoti hai    |

### **Constant**
- #### Constant ka matlab:
    - #### Ek fixed value jo program ke har jagah use ki ja sakti hai, aur kabhi change nahi hoti

<br>

- #### Constant hamesha immutable hoti hai
- #### Type explicitly define karna zaroori hai
- #### Constant ka naam uppercase letters me likha jata hai (ye Rust convention hai)
- #### Scope program-wide ho sakta hai

```rust
fn main(){
    const PI: f64 = 3.14159;
    println!("Value of pi is {}",PI);
}
```

```rust
fn main(){
    const PI: f64 = 3.14159;

    // try to change const value
    const PI:f64 = 2.5;
}
```

### Scope ka matlab:
 - #### Wo area jahan variable accessible hota hai.
 - #### Rust me: har variable ka ek block-level scope hota hai ({ } braces ke andar).

<body style="font-family: Sans-serif;">
    <div style="color: black; font-size: 15px; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black; padding: 20px;">

- #### Block Scope (Local Scope):
    - #### Jo variable { } braces ke andar define hota hai, wo sirf usi block me accessible hota hai


```rust
fn main() {
    let x = 2;
    {
        let x = 10; // x ka scope sirf is block me
        println!("{}", x); // ✔ 10
    }

    
    println!("{}", x);
}
```

<body style="font-family: Sans-serif;">

- #### Rust me har variable ka scope block-level hota hai: jo { } ke andar banega, sirf usi block me accessible hoga; bahar access nahi ho sakta
- #### { } ke andar jo naya x hai → shadowing + block-level scope
- #### { } ke bahar outer x ki value wahi rahegi

```rust
fn main() {
    let y = 2;
    {
        let x = 10; 
        println!("value of x inner bracket is {}", x);
        println!("value of y from outer bracket is {}", y);
    }

    println!("value of y is {}", y);
}
```

- ### Local Variable:
    - #### Wo variable jo sirf ek function ya block ke andar accessible hota hai.
    - #### Scope limited hota hai
    - #### Block/Function ke bahar use nahi kar sakte
    - #### Memory automatically free ho jati hai jab block/function khatam hota hai

```rust
fn main() {
    let x = 10; // local variable
    println!("{}", x); // ✔ accessible
}
```

- ### Global Variable:
    - #### Wo variable jo program ke har jagah accessible ho, aur poore runtime me memory me rehta ho.
    -  #### Rust me direct global let variables nahi hote
    -  #### Global ke liye static ya const use karte hain

```rust
const PI: f64 = 3.14159;

fn main() {
    println!("----- Main Function ------");
    println!("{}", PI);
    other_func();
}

fn other_func() {
    println!("----- Other Function ------");
    println!("{}", PI); 
}
```

### **Static Variable**

- #### static ek global variable hota hai
- #### Program ke poore runtime me memory me rehta hai
- #### Default immutable hai
- #### Agar mut lagao → mutable ho jata hai, lekin unsafe block me hi access kar sakte ho

```rust
static pi :f64 = 3.14159;

fn main() {
    println!("----- Main Function ------");
    println!("{}", pi);
    other_func();
}

fn other_func() {
    println!("----- Other Function ------");
    println!("{}", pi); 
}
```

<h2 style="text-align:center;">static vs const</h3>

| Feature            | `const`                    | `static`                                                       |
| ------------------ | -------------------------- | -------------------------------------------------------------- |
| **Mutability**     | Hamesha immutable          | Default immutable, `mut` lagake mutable bana sakte ho          |
| **Change Value**   | ❌ Kabhi change nahi hoti   | ✔ Mutable static ke liye `unsafe` block me change ho sakti hai |
| **Scope**          | Global ya block-level      | Hamesha global (program-wide)                                  |
| **Memory**         | Compile-time constant      | Fixed memory during runtime                                    |
| **Thread Safety**  | Safe                       | Mutable static unsafe → thread-safe nahi by default            |
| **Initialization** | Compile-time required      | Compile-time required                                          |


```rust
static mut COUNTER: u32 = 0;
fn main() {
    unsafe {
        COUNTER += 1; // allowed
        println!("Counter = {}", COUNTER);
    }
}
```
