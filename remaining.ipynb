# pointer refferences

#### Ek time pe 1 mutable reference allowed

````rust
fn main(){
    let mut x = 10;
    let a = &mut x;   // ok
    // let b = &mut x;   // error
}
````

### Multiple immutable references allowed

````rust
fn main(){

    let x = 10;
    let a = &x;
    let b = &x;   // ok
}
````

#### Immutable + mutable together allowed nahi

````rust
fn main(){
let mut x = 10;
let a = &x;    
let b = &mut x; // error
}
main();
````


# ‚úÖ **1) Ownership (Sabse main rule)**

Rust ka sabse bara rule:

> **Har value ka sirf 1 owner hota hai.**

Example:

```rust
let s = String::from("hello");
let t = s;
```

Yahan:

* `t` owner ban gaya
* `s` **invalid** ho gaya
* Use karne pe error

---

# ‚úÖ **2) Move rule**

Agar aap variable ko assign ya function mein pass karte ho:

```rust
let a = String::from("abc");
let b = a;  // move
```

* Ownership move hoti hai
* Copy nahi hoti (except Copy types)

---

# ‚úÖ **3) Copy rule**

Primitive types automatically copy hote hain:

* integers
* floats
* bool
* char
* fixed-size arrays

Example:

```rust
let x = 5;
let y = x;   // copy, x still valid
```

---

# ‚úÖ **4) Borrowing Rules (Bahut important)**

Borrowing 2 types ka hota hai:

## üîπ **(A) Immutable Borrow**

```rust
let s = String::from("hello");
let a = &s;
let b = &s;   // allowed
```

‚úî multiple immutable borrows allowed

---

## üîπ **(B) Mutable Borrow**

```rust
let mut s = String::from("hello");
let a = &mut s;
```

‚ùå Multiple mutable borrows **allowed nahi**

---

# üü• **5) Mut + Immutable sath allowed nahi**

```rust
let mut s = String::from("hello");

let a = &s;    // immutable
let b = &mut s; // ‚ùå error
```

Reason: Data race prevent karna.

---

# üü¢ **6) Ek time pe ya to:**

* multiple **immutable** references
  **OR**
* ek hi **mutable** reference

---

# üü† **7) Borrow lifetime rule**

Borrowed reference tab tak valid hota hai jab tak:

* owner valid ho
* mutation ke saath conflict na ho

Rust compile time pe check karta hai.

---

# üü£ **8) Shadowing vs Mutation**

## üîπ Mutation (`mut`)

Same variable ki value change hoti hai:

```rust
let mut x = 10;
x = 20;
```

Type same rehega.

---

## üîπ Shadowing (`let x = ...`)

Naya variable same naam se ban jata hai:

```rust
let x = 10;
let x = "text";   // allowed
```

‚úî Type change allowed
‚úî Immutable se phir immutable allowed
‚úî Mut ki zaroorat nahi

---

# üî• Final Summary (Super Short)

| Feature          | Rule                              |
| ---------------- | --------------------------------- |
| Ownership        | Har value ka 1 owner              |
| Move             | Owner change, old invalid         |
| Copy             | Primitive types copy              |
| Borrow           | Reference dena                    |
| Immutable borrow | unlimited                         |
| Mutable borrow   | only one                          |
| Mut + Immutable  | same time allowed nahi            |
| Shadowing        | new variable, type change allowed |
| Mutation         | same variable value change        |
