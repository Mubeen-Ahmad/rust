<body style="font-family: Sans-serif;">

### **What is Identifers**

- #### identifiers wo names hote hain jo aap apne code ke andar cheezon ko pehchanane (identify) ke liye use karte ho.
- #### Means jo naam aap dete ho:
    - #### variable 
    - #### function
    - etc

<body style="font-family: Sans-serif;">
    <div style="color: black; font-size: 15px; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black; padding: 20px;">

### Variable identifier

```rust
let x = 10;
```

`x` → identifier

### Function identifier

```rust
fn display() {}
```

`display` → identifier

### Struct identifier

```rust
struct Person {}
```

`Person` → identifier

### Module identifier

```rust
mod tools {}
```

`tools` → identifier

### Enum identifier

```rust
enum Color { Red, Blue }
```

`Color`, `Red`, `Blue` → identifiers

### Constant identifier

```rust
const MAX: i32 = 100;
```

`MAX` → identifier


**So identifier wo naam hai jo code ke kisi bhi item ko identify karta hai.
Jab aap kisi variable, function, struct, enum ko naam dete ho, us naam ko identifier kehte hain.**

<h1 style="text-align:center; color:#005bbd; font-size:20px; font-family:Sans-serif; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px Black;">
   Identifier Rules
</h1>

<body style="font-family: Sans-serif;">
    <div style="color: black; font-size: 15px; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black; padding: 20px;">

- #### 1) Sirf yeh characters allowed hain
    - #### Alphabets: A–Z, a–z
    - #### Digits: 0–9
    - #### Underscore: _

````rust
fn My_Function_Identifer(){

    let my_variable_identifer = 10;
}
````

<body style="font-family: Sans-serif;">
    <div style="color: black; font-size: 15px; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black; padding: 20px;">

- #### 2) Identifier digit se start nahi ho sakta
    - #### ✔ Alphabet se start ho sakta hai
    - #### ✔ Underscore se start ho sakta hai
    - #### ❌ Number se start nahi hoga

````rust
let age = 10;   // ok
let _value = 5; // ok
````

````rust
let 1num = 3;   // ❌ not allowed
````

<body style="font-family: Sans-serif;">
    <div style="color: black; font-size: 15px; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black; padding: 20px;">

- #### 3) Spaces Not allowed

````rust
let first_name = "Ali"; // ok
````

````rust
let first name = "Ali"; // ❌ error
````

<body style="font-family: Sans-serif;">
    <div style="color: black; font-size: 15px; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black; padding: 20px;">

- #### 4) Spaces Not allowed
    - #### Case-sensitive language
      - #### e.g
          - #### name
          - #### Name
          - #### NAME
      - #### ye teen alag identifiers hain.

````rust
fn main(){

    let name = "a";
    let Name = "b";
    let NAME = "c";

    println!("{} {} {}",name,Name,NAME);
}
````

<body style="font-family: Sans-serif;">
    <div style="color: black; font-size: 15px; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black; padding: 20px;">

- #### 5) keywords identifier nahi ban sakte

````rust
fn main(){
    let fn = 10;   // ❌ not allowed
}
````

<body style="font-family: Sans-serif;">
    <div style="color: black; font-size: 15px; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black; padding: 20px;">

- #### 6) Unicode identifiers allowed (Rust , python feature)
    - ####  lekin recommend nahi.

````rust
fn main(){
    let π = 3.14;
    println!("Value of π is {}",π);
}
````

<body style="font-family: Sans-serif;">
    <div style="color: black; font-size: 15px; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black; padding: 20px;">

- #### 7) _ (single underscore) ka special meaning
    - ####  _ ek throw-away identifier hai
    - #### isko compiler ignore karta hai.

````rust
fn main(){
    let _ = 10; // result ignore

    println!("{}",_); // error
}
````

<body style="font-family: Sans-serif;">
    <div style="color: black; font-size: 15px; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black; padding: 20px;">

- #### 8) __ Multiple leading underscores allowed
  

````rust
fn main(){
    let __ = 10;
    println!("{}",__); 
}
````

<body style="font-family: Sans-serif;">
    <div style="color: black; font-size: 15px; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black; padding: 20px;">

- #### 9) Identifiers mein symbols allowed nahi
  

````rust
fn main(){
    let name@ = 10;
    println!("{}",name@); 
}
````

<h1 style="text-align:center; color:#005bbd; font-size:20px; font-family:Sans-serif; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px Black;">
   Multiple Variables in Single Line
</h1>

<body style="font-family: Sans-serif;">
    <div style="color: black; font-size: 15px; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black; padding: 20px;">

#### **Rust mein ek hi line mein multiple variables declare + assign karna allowed hai aur bohot simple hai.**

<body style="font-family: Sans-serif;">
    <div style="color: black; font-size: 15px; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black; padding: 20px;">

- #### 1. Multiple variables ek line mein declare + assign

````rust
fn main(){
    let (x, y, z) = (10, 20, 30);
    println!("{} {} {}",x,y,z);
}
````

<body style="font-family: Sans-serif;">
    <div style="color: black; font-size: 15px; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black; padding: 20px;">

- #### 2. Multiple variables declare (baad mein assign karna ho to)

````rust
fn main(){
    let (x, y);
    x = 10;
    y = 20;
    println!("{} {}",x,y);
}
````

<body style="font-family: Sans-serif;">
    <div style="color: black; font-size: 15px; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black; padding: 20px;">

- #### 3. Swap Variable

````rust
fn main(){
    let (x, y) = (2 , 5);
    println!("x = {x} y = {y}",x=x,y=y);
    
    println!("SWAP");
    let (x,y) = (y,x);
    println!("x = {x} y = {y}",x=x,y=y);
}
````
